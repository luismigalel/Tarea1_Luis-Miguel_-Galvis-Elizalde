##4. Verifique que se cumple el teorema de Parseval (reporte el resultado en W). Tenga en cuenta que la información que inicialmente se le dio corresponde a campo óptico (raíz cuadrada de la potencia lumínica por unidad de área [(W/𝜇m^2)^0.5]).

Ahora comprobemos que se cumple el Teorema de parseval y que obtenemos igual cantidad de potencia en la siguiente igualdad:


$$ \int \mid f(t) \mid^2 \text{d} x \text{d} y \ \color{red}{} = \int \mid f(\nu) \mid^2 \text{d} \nu\text{d} \nu \ \color{red}{} $$

```python
dxy = (0.042)**2 #Diferencial de area.
###########################


# Tranformada de Fourier: espectro de frecuencias de la imagen.

TF_IM1 = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(IM1)))*dxy




IM1FI=IM1**2
TFFI=TF_IM1
sd=sum(abs((IM1FI)**2))
kd=sum(abs((TFFI)**2))


# Cálculo de la  total en cada espacio, las unidades son Watts.
suma1= sum((sd))*dxy                     #Integral del lado izquierdo.

suma3 = sum(kd)*1/(((3072)**2)*dxy)      #Integral del lado derecho.

print('Potencia en plano de entrada = %f' % suma1, "W")
print('Potencia en plano de salida  = %f' % suma3, "W")

print("Hemos comprobado así que el teorema de Parseval se cumple y la energía")
```
